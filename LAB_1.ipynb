```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y1=[0,10,40,30,20]
y2=[0,20,30,30,40]
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("Year")
plt.ylabel("Growth Rate")
plt.title("Growth rate of companies")
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y1=[0,10,40,30,20]
z1=[15,35,55,40,25]
y2=[23,43,53,13,14]
z2=[16,26,10,36,46]
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot(x,y1,z1, label = "3D curve")
ax.plot(x,y2,z2, label = "3D curve")
```
